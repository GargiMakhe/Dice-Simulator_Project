```python
import random
from IPython.display import clear_output

# Colors (ANSI)
class color:
    RED = '\033[91m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    BLUE = '\033[94m'
    END = '\033[0m'

# Dice Faces
dice_faces = {
    1: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ         в”Ӯ\nв”Ӯ    в—Ҹ    в”Ӯ\nв”Ӯ         в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ",
    2: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ  в—Ҹ      в”Ӯ\nв”Ӯ         в”Ӯ\nв”Ӯ      в—Ҹ  в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ",
    3: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ  в—Ҹ      в”Ӯ\nв”Ӯ    в—Ҹ    в”Ӯ\nв”Ӯ      в—Ҹ  в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ",
    4: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв”Ӯ         в”Ӯ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ",
    5: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв”Ӯ    в—Ҹ    в”Ӯ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ",
    6: "в”Ңв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”җ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв”Ӯ  в—Ҹ   в—Ҹ  в”Ӯ\nв””в”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”Җв”ҳ"
}

# Data storage
roll_history = []
frequency = {i: 0 for i in range(1, 7)}

# Fast roll (no delay, minimal clear)
def roll_animation():
    clear_output(wait=True)
    print(color.YELLOW + "рҹҺІ Rolling Dice...\n" + color.END)
    print(dice_faces[random.randint(1, 6)])

# Show history
def show_history():
    print(color.BLUE + "\nрҹ“ң Roll History:\n" + color.END)

    if not roll_history:
        print("No rolls yet.")
        return

    print("No.   Dice   Value")
    print("----------------------")
    for i, val in enumerate(roll_history, 1):
        print(f"{i:<5} {dice_faces[val].splitlines()[2].strip()}    {val}")

# Show statistics (graph kept)
def show_stats():
    print(color.GREEN + "\nрҹ“Ҡ Statistics:\n" + color.END)
    print(f"Total Rolls: {len(roll_history)}\n")

    for i in range(1, 7):
        print(f"{i}: {frequency[i]} times")

    choice = input("\nShow graph? (y/n): ").lower()
    if choice == 'y':
        import matplotlib.pyplot as plt

        x = list(frequency.keys())
        y = list(frequency.values())

        plt.figure()
        plt.bar(x, y)
        plt.xlabel("Dice Face")
        plt.ylabel("Frequency")
        plt.title("Dice Roll Statistics")
        plt.show()
        plt.close()   # prevents lag after graph

# Main simulator
def dice_simulator():
    print(color.GREEN + "рҹҺІ Advanced Dice Rolling Simulator рҹҺІ" + color.END)

    while True:
        try:
            choice = input("\nEnter 'r' to roll or 'q' to quit: ").lower().strip()

            if choice == 'r':
                roll_animation()
                result = random.randint(1, 6)

                roll_history.append(result)
                frequency[result] += 1

                print(color.GREEN + "\nрҹҺү You rolled:\n" + color.END)
                print(dice_faces[result])
                print(f"\nValue: {result}")

                while True:
                    print("\nWhat would you like to do next?")
                    print("1. Roll Again")
                    print("2. View History")
                    print("3. View Statistics")
                    print("4. Quit")

                    sub_choice = input("Enter choice (1-4): ").strip()

                    if sub_choice == '1':
                        break
                    elif sub_choice == '2':
                        show_history()
                    elif sub_choice == '3':
                        show_stats()
                    elif sub_choice == '4':
                        print("рҹ‘Ӣ Exiting... Goodbye!")
                        return
                    else:
                        print(color.RED + "вҡ пёҸ Invalid choice!" + color.END)

            elif choice == 'q':
                print("рҹ‘Ӣ Exiting... Goodbye!")
                break

            else:
                raise ValueError("Please enter only 'r' or 'q'.")

        except ValueError as ve:
            print(color.RED + "вҡ пёҸ " + str(ve) + color.END)

        except Exception as e:
            print(color.RED + "вқҢ Unexpected Error: " + str(e) + color.END)

# Run
dice_simulator()
```
